```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'math_score': [20, 30, 40, 90],
        'english_score': [70, 80, 90, 90],
        'history_score': [50, 60, 70, 90]}

# Create a DataFrame using Pandas
df = pd.DataFrame(data)

# 1. Calculate the average score for each student
# Pandas is used to calculate the mean of columns and create a new column 'average_score'
df['average_score'] = df[['math_score', 'english_score', 'history_score']].mean(axis=1)

# 2. Find the student with the highest total score
# Pandas is used to locate the row with the maximum average score
highest_scoring_student = df.loc[df['average_score'].idxmax()]

# 3. Identify students who need improvement (average score below 80)
# Pandas is used to filter students with average scores below 80
students_needing_improvement = df[df['average_score'] < 80]

# Display the refined results without including the index
print("Student Performance Analysis:")
print("============================\n")

# 1. Calculate the average score for each student
# Pandas is used to display the average score for each student in a tabular format
print("1. Average Score for Each Student:")
print(df[['name', 'average_score']].to_string(index=False))
print("\n")

# 2. Find the student with the highest total score
# Pandas is used to display the student with the highest average score
print("2. Student with the Highest Total Score:")
print(highest_scoring_student[['name', 'average_score']].to_string(index=False))
print("\n")

# 3. Identify students who need improvement (average score below 80)
# Pandas is used to display students needing improvement based on the average score
print("3. Students Needing Improvement:")
print(students_needing_improvement[['name', 'average_score']].to_string(index=False))
print("\n")

# Visualizations

# Bar chart for Average Scores with student names using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(df['name'], df['average_score'], color='lightcoral')
plt.title('Average Scores for Each Student')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.ylim(0, 100)

# Displaying the average score on each bar
for i, v in enumerate(df['average_score']):
    plt.text(i, v + 1, str(round(v, 2)), ha='center', va='bottom', fontsize=8)

# Show the plot
plt.show()
```
